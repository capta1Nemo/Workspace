```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Set a seed for reproducibility
np.random.seed(0)

# 2. Generate random features
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = np.random.rand(n_samples)

# 3. Create a linear relationship (true underlying model)
# y = 2.5 * x1 + 1.5 * x2 + 1.0 (intercept term)
true_coefficients = [1.0, 2.5, 1.5]
y = true_coefficients[0] + true_coefficients[1] * x1 + true_coefficients[2] * x2

# 4. Add noise to the target variable
noise = np.random.randn(n_samples) * 0.1
y_noisy = y + noise

# 5. Combine features and target into a DataFrame (optional)
data = pd.DataFrame({
    'x1': x1,
    'x2': x2,
    'y': y_noisy
})

# Display the first few rows of the dataset
print(data.head())

# Plot the data to visualize the relationships
plt.figure(figsize=(12, 5))

# Plot x1 vs y
plt.subplot(1, 2, 1)
plt.scatter(x1, y_noisy)
plt.xlabel('x1')
plt.ylabel('y')
plt.title('x1 vs y')

# Plot x2 vs y
plt.subplot(1, 2, 2)
plt.scatter(x2, y_noisy)
plt.xlabel('x2')
plt.ylabel('y')
plt.title('x2 vs y')

plt.tight_layout()
plt.show()
```

linear regression for classification

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generate the synthetic dataset
np.random.seed(0)
n_samples_per_class = 50

# Class +1
x1_class1 = np.random.rand(n_samples_per_class) * 0.5 + 0.5
x2_class1 = np.random.rand(n_samples_per_class) * 0.5 + 0.5
y_class1 = np.ones(n_samples_per_class)

# Class -1
x1_class2 = np.random.rand(n_samples_per_class) * 0.5
x2_class2 = np.random.rand(n_samples_per_class) * 0.5
y_class2 = -np.ones(n_samples_per_class)

# Combine the data
x1 = np.concatenate([x1_class1, x1_class2])
x2 = np.concatenate([x2_class1, x2_class2])
y = np.concatenate([y_class1, y_class2])

# Create the DataFrame
data = pd.DataFrame({'x1': x1, 'x2': x2, 'y': y})

# Prepare the data for linear regression
X = data[['x1', 'x2']]  # Features
y = data['y']            # Target

# Train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Get the coefficients (weights) and intercept
coef = model.coef_
intercept = model.intercept_

# Define the decision boundary function
def decision_boundary(x1, coef, intercept):
    return -(coef[0] * x1 + intercept) / coef[1]

# Plot the data points and the decision boundary
plt.figure(figsize=(8, 6))
plt.scatter(x1_class1, x2_class1, color='blue', label='Class +1')
plt.scatter(x1_class2, x2_class2, color='red', label='Class -1')

# Calculate the decision boundary line
x1_range = np.linspace(0, 1, 100)
x2_boundary = decision_boundary(x1_range, coef, intercept)

# Plot the decision boundary
plt.plot(x1_range, x2_boundary, color='green', linestyle='--', label='Decision Boundary')

# Labels and legend
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Linear Regression Decision Boundary')
plt.legend()
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Define the plane equation: z = 2x + 3y
z = 2 * x + 3 * y

# Create a new figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, alpha=0.5, rstride=100, cstride=100, color='yellow')

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Set title
ax.set_title('Plane: z = 2x + 3y')

# Show plot
plt.show()
```
