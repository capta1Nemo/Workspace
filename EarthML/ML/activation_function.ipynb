```python
from matplotlib import pyplot as plt
import numpy as np
```

activation functions


```python
inputs = [x for x in range(-10, 10)]
```

hidden layers:reLU,tanh, sigmoid

-Multilayer Perceptron (MLP): ReLU activation function.

-Convolutional Neural Network (CNN): ReLU activation function.

-Recurrent Neural Network: Tanh and/or Sigmoid activation function.

```python
def reLU(x):
    return max(0.0, x)

outputs = [reLU(x) for x in inputs]
plt.plot(inputs, outputs)
plt.show()
```

```python

def sigmoid(x):
 return 1.0 / (1.0 + np.exp(-x))

outputs = [sigmoid(x) for x in inputs]
plt.plot(inputs, outputs)
plt.show()
```

```python
def tanh(x):
 return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
 
outputs = [tanh(x) for x in inputs]
plt.plot(inputs, outputs)
plt.show()
```

output layers:linear,sigmoid,softmax

-Regression: One node, linear activation.

-Binary Classification: One node, sigmoid activation.

-Multiclass Classification: One node per class, softmax activation.

-Multilabel Classification: One node per class, sigmoid activation.

```python
#linear gives same x
outputs = [x for x in inputs]
# plot inputs vs outputs
plt.plot(inputs, outputs)
plt.show()
```

softmax is probability distribution of each input based on its value and calculated as such : e^x / sum(e^x)

```python
def softmax(x):
    return np.exp(x) / np.exp(x).sum()
#new input
inputs = [1.0, 3.0, 2.0]
outputs = softmax(inputs)
print(outputs)
print(outputs.sum())
```

eklenti (overfitt örneğinde gördüm)

```python
def eLU(x,alpha=1.0):
    if x > 0:
        return x
    else: return alpha*np.exp(x) - 1
```

```python
inputs = [x for x in range(-10, 10)]
outputs = [eLU(x) for x in inputs]
plt.plot(inputs, outputs)
plt.show()
```
