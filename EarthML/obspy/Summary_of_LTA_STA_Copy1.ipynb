## Introduction to Summary of LTA STA Report

- **In LTA-STA report, our purpose was showing the effects of some values which are threshold and window-size value. These comparisons are done by two main packages that are ObsPy and EQcorrscan. We tried seven different threshold values and window-size values, these values consist of two other values which are lower and higher value.**

## The Definitions of Parameters

* **nsta: Length of short time average window in samples**
* **nlta: Length of long time average window in samples**
* **tr.data: NumPy array**
- **Both are used in the main code which is "cft = recursive_sta_lta(tr1.data, int(2*df),int(df*60))"**
   

* **thres1: Value above which trigger (of characteristic function) is activated (higher threshold)**
* **thres2: Value below which trigger (of characteristic function) is deactivated (lower threshold)**
- **Both are usef in the second main code that is "on_of = trigger_onset(cft, 0.5, 0.25)"**

- **There are also parameters for EQcorrscan, it is too similar to ObsPy's parameters:**
   * **stalen: Length of the short-term average window in seconds.**
   * **ltalen: Length of the long-term average window in seconds.**
   * **trig_on: STA/LTA ratio to trigger a detection/pick**
   * **trig_off: STA/LTA ratio to turn the trigger off - no further picks will be made between exceeding trig_on until trig_off is reached.**
   - **These are used in that code "events=eqcorrscan.utils.picker.stalta_pick(st,  stalen=50, ltalen=5000, trig_on=2.75, trig_off=1.75)"**

- **Also, we wanted to show you the preprocessing effect, removing mean and applying taper. And, once we did preprocess, we got better results than raw(nothing applied) data. All these were shown our whole report, but in that summary, we will indicate just the best models with photographs. Moreover, all these results are compared with the catalog, there are some Pandas codes.**

## Some Inferences

- **After the lots of codes, we have some inferences:**
   - **As the threshold value increases the number of detection is decreased especially the model loses its efficiency on the weak earthquakes, but as you read the biggest earthquake is detected though.**
   
   - **Again, as the window-size value increases the number of detection is again decreased. We can assimilate it as threshold effect. But, as you saw, the big earthquake(5.8) is always detected.**
   
   - **Eqcorrscan package maybe is better than ObsPy since the source code of the Eqcorrscan showed us that it makes some elimination on the detections. But, again we want to say that Eqcorrscan module gives us the detection times twice. It is bad for us but, the detections, that are found by Eqcorrscan, make more sense than ObsPy.**
   
   - **Tapering and removing mean is beneficial for us, since they prevent deconvolution. And, this circumstance makes the model better. Also, once we use taper and remove the mean, the number of detection is increased. But, as you saw the best model worked on Eqcorrscan(mean removed & taper applied).**
