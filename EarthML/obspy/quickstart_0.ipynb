### **Sismolojik Verilerle İlgili Kavramlar**
Burada yazı serimiz boyunca karşılaşacağınız bazı kavramların açıklaması yer almaktadır. Kavramları türkçeye çevirdik ancak her zaman Türkçe karşılıklar kullanılmayacak. Burada olmayan bazı kavramlar yeri geldikçe yazılar içerisinde açıklanacak, olan kavramlara dair ek açıklamalar yeri geldikçe yapılacaktır

><mark>**Olay (Event)**</mark>: Bir deprem veya sismik olayı ifade eder. Konum, derinlik ve büyüklük gibi detayları içerir.

><mark>**Ağ (Network)**</mark>: Geniş bir alan üzerinde deprem aktivitesini izlemek amacıyla birbirine bağlı sismoloji istasyonlarının oluşturduğu bütüne denir. Sismoloji ağları, depremlerin erken uyarısı sağlamak ve araştırma için veri toplamak için kullanılır. Kandilli Rasathanesi'nin Ağ kısaltması sonraki adımlarda da göreceğimiz üzere `KO` dur.

><mark>**İstasyon (Station)**</mark>: Sismometre ve diğer enstrümanlarla yer hareketini izlemek ve sismik veri kaydetmek için kullanılan bir gözlem ve ölçüm noktasıdır. Sismik bir olay, genellikle birden çok istasyon tarafından kaydedilir.

><mark>**Büyüklük (Magnitude)**</mark>: Bir depremin serbest bıraktığı enerjinin ölçüsüdür. En yaygın büyüklük ölçeği [Richter ölçeği](https://tr.wikipedia.org/wiki/Richter_%C3%B6l%C3%A7e%C4%9Fi) veya [moment büyüklüğü ölçeği](https://tr.wikipedia.org/wiki/Moment_magnit%C3%BCd_%C3%B6l%C3%A7e%C4%9Fi) (Mw) dir.

><mark>**Katalog (Catalog)**</mark>: Sismik olayların tarih, saat, konum ve büyüklük gibi deprem detaylarını içeren bir veri tabanı veya kayıttır.

><mark>**Envanter (Inventory)**</mark>: Sismoloji istasyonlarında kullanılan ekipman ve cihazların listesidir. Bu ekipmanlar arasında sismometreler, veri kaydedicileri, bilgisayarlar, çeşitli sensörler vs. bulunur.

><mark>**Kanal (Channel)**</mark>: Sismolojide kanal, sismik dalgaları tespit etmek ve kaydetmek için kullanılan belirli bir kayıt cihazı veya sensörü ifade eder. Farklı yönlerdeki yer hareketlerini yakalamak için genellikle birden fazla kanal kullanılır (Bunu daha sonra tekrar açıklayacağız).  

><mark>**Zaman Serisi (Time Series)**</mark>: [Zaman serisi](https://tr.wikipedia.org/wiki/Zaman_serisi), bir belirli ölçümün veya olayın zaman içindeki değişimini temsil eden veri dizisidir. Sismolojik veriler için zaman serileri genelde zaman bilgisi ve genlik bilgisini içerir.
  
><mark>**Stream (Veri Akışı)**</mark>: Bir sismoloji istasyonundaki bir veya daha fazla kanaldan gelen sürekli sismik veri akışını ifade eder.

><mark>**İz (Trace)**</mark>: Sismolojide iz, yer hareketini ölçerken tek bir sismograf tarafından kaydedilen eğriyi tanımlar. 'İz' ismi, bir sismografın kağıt rulosu döndükçe iğnenin bıraktığı izden gelir.   
Günümüzde cihazlar veriyi dijital olarak kaydeder, ve iz dijital bir eğriye karşılık gelir.   
`Trace`ler `Stream`in bir parçasıdır, bir `Stream`de bir veya birden fazla `Trace` bulunabilir.

><mark>**Dalga Formu (Waveform)**</mark>: Bir sismometre tarafından kaydedilen zaman içinde yer hareketinin değişimini gösteren grafik bir temsildir. Dalga formları, sismik dalgaların varışını ve karakteristiklerini gösterir.  

><mark>**Zaman Aralığı (Time Window)**</mark>: Bir başlangıç noktasından bir bitiş noktasına kadar geçen süreyi tanımlar.

><mark>**MiniSEED/.mseed**</mark>: Sismolojide dijital sismik verileri depolamak ve paylaşmak için yaygın olarak kullanılan bir veri formatıdır. MiniSEED ve diğer sismolojik veri formatlarıyla alakalı bilgi için [buraya tıklayın.](http://www.ds.iris.edu/ds/nodes/dmc/data/formats/)

## 4. Deprem Verilerine ObsPy ile Bir Bakış

`obspy` kütüphanesindeki `Stream` nesnesi, sismolojik verileri depolamak ve işlemek için kullanılan bir veri yapısıdır.  genellikle birden çok `Trace` nesnesini içerir, bu `Trace` nesnelerinin her biri bir istasyonun belirli bir bileşeni için zaman serisi verilerini temsil eder. `Stream` nesnesi, bir veya daha fazla istasyonun sismik izlerini içerir.  

İndirdiğimiz veriyi bir 'Stream' nesnesi olarak `obspy` ile okuyalım.   

```python
st = read(r"C:\Users\sukru\OneDrive\Desktop\Workspace\EarthML\Package.mseed") #Deprem verilerini içeren .mseed dosyasının konumunu kopyalayın.
print(st)
```

Tüm `Trace` isimleri:  

>`AĞ_İSMİ.İSTASYON_İSMİ. .TRACE_KODU`  

Formatına sahiptir. `Trace` kodları, sismometredeki bileşenlerin yönü gibi bilgileri verir.

`obspy`daki `Stream` nesnesi `python`daki `list`e benzerdir, tıpkı bir listede yaptığımız gibi, `Stream` içindeki `Trace`lerin her birine indeks kullanarak erişebiliriz.

`Trace` nesneleri, zaman serisi verileri haricinde verinin geldiği ağ, verinin başlangıç zamanı, veri formatı gibi bazı meta bilgileri de içerir. Bunlara `Trace.stats` özniteliğiyle (attribute) ile erişebiliriz.  

```python
for i in range(len(st)):
    tr = st[i] #Tracelerden bir tanesini bu değişkene atayalım.
    print(f"Trace {i+1}: \n",tr)
    print(f"Trace {i+1} Stats: \n",tr.stats)
    print("\n")
print(st)    
```

Bu `Trace.stats`ın bize sunduğu bilgilere ayrı ayrı da istediğinizde erişebilirsiniz.

```python
tr1=st[0]
print(tr1.stats.starttime)
print(tr1.stats.network)
```

## Basit Görselleştirme

`obspy`, dalga formlarını görselleştirmemiz için fonksiyonlar bulundurur. Daha sonra bunu detaylı olarak göreceğiz ancak şimdilik elimizdeki veriyi basitçe görselleştirelim.  

Daha önce tanımladığımız `trace`in (tr1) grafiğini çizdirelim (`plot`).

```python
tr1.plot(starttime=tr1.stats.starttime-10, endtime=tr1.stats.endtime+10,color='red',bgcolor='black',linewidth='0.2');
```

Görüldüğü üzere `Trace` objesi birçok parametre alır. Bunlardan bazıları grafiğin görselliğini değiştirirken göstermediğimiz bazıları ölçeğini değiştirmeye yarar.  

Plot fonksiyonu ile ilgili daha fazla bilgiyi ObsPy dökümantasyonundan [edinebilirsiniz](https://docs.obspy.org/packages/autogen/obspy.core.trace.Trace.plot.html)
